```python
import numpy as np
import pandas as pd
```

```python
data = np.array([[0,96],
                [1,85],
                [2,82],
                [3,74],
                [3,95],
                [5,68],
                [5,76],
                [5,84],
                [6,58],
                [7,65],
                [7,75],
                [10,50]])
df = pd.DataFrame(data,columns=['Hour','marks'])
x=df['Hour']
y=df['marks']
```

```python
n=len(x)
sum_x= np.sum(x)
sum_y = np.sum(y)
sum_xy = np.sum(x*y)
sum_x2 = np.sum(x*x)
m = (n*(sum_xy) - (sum_x*sum_y))/((n*sum_x2)-(sum_x**2))
c = (sum_y - m*sum_x)/n
```

```python
print("slope:",m)
print("intercept:",c)
```

```python
pred_hour = int(input("enter the no of hour "))
def predict(hour):
    return c + m * hour
if(pred_hour>=0 and pred_hour<=10):
    print(f"predicted marks for {pred_hour} hours study",predict(pred_hour))
else:
    print("please enter values between 1 to 10")
    
```
